```python
## Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Loading the data
data=pd.read_csv('HR-Employee-Attrition.csv')
```

```python
## Getting all columns form the dataset
data.columns
```

## Conversion of Categorical variables

```python
## Categorical data conversion
data.head()
```

```python
data.Attrition.unique()
#checking unique value in Attrition colum
```

```python
## Manual encoding Attrition feature
data.Attrition=data.Attrition.map({'Yes':0,'No':1})
data
```

```python
data.BusinessTravel=data.BusinessTravel.map({'Travel_Frequently':0,'Travel_Rarely':1,'Non-Travel':2})
data
```

```python
data.Department=data.Department.map({'Research & Development':0,'Sales':1,'Human Resources':2})#imputation using map function
```

```python
data.head()
```

```python
data.EducationField=data.EducationField.map({'Life Sciences':0,'Medical':1,'Marketing':2,'Technical Degree':3,'Other':4,'Human Resources':5 })
```

```python
data.head()
```

```python
data.Gender.value_counts()
```

```python
data.Gender=data.Gender.map({'Male':0,'Female':1})
data
```

```python
#data.Gender=pd.get_dummies(data.Gender,drop_first=True)
```

```python
data.head()
```

```python
pd.set_option("display.max_columns", None)
```

```python
data.head()
```

```python
data.JobRole=data.JobRole.map({'Laboratory Technician':0,'Sales Executive':1,'Research Scientist':2,'Sales Representative':3,
                              'Human Resources':4,'Manufacturing Director':5,'Healthcare Representative':6,'Manager':7,'Research Director':8 })
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder#importing label encoder from sklearn 

label = LabelEncoder()#object creation 
data.MaritalStatus=label.fit_transform(data.MaritalStatus)
```

```python
data.head()
```

```python
data.OverTime=label.fit_transform(data.OverTime)
```

```python
data.drop(["Over18"], axis = 1, inplace = True)
```

```python
data.head()
```

# Model Creation

```python
## Creating independent and dependent variable
X = data.drop('Attrition', axis=1)#independent variable 
y = data.Attrition#dependent variable 
```

```python
data.head()
```

```python
## Balacing the data
from collections import Counter# importing counter to check count of each label
from imblearn.over_sampling import SMOTE #for balancing the data
sm=SMOTE()#object creation
print(Counter(y))# checking count for each class 
X_sm,y_sm=sm.fit_resample(X,y)#applying sampling on target variable 
print(Counter(y_sm))# checking count after sampling for  each class
```

```python
## preparing training and testing data
from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.25, random_state=42)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier#importing randomforest

rf_clf = RandomForestClassifier(n_estimators=100)#object creation ,taking 100 decision tree in random forest 
rf_clf.fit(X_train,y_train)#training the data
```

```python
y_predict=rf_clf.predict(X_test)#testing
```

```python
from sklearn.metrics import f1_score
f_Score=f1_score(y_test,y_predict)
f_Score
```

# Hyperparameter Tunning

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]#List Comprehension-using for loop in list
max_features = ['auto', 'sqrt']#maximum number of features allowed to try in individual tree
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]#List Comprehension-using for loop in list
max_depth.append(None)
min_samples_split = [2, 5, 10]#minimum number of samples required to split an internal node
min_samples_leaf = [1, 2, 4]#minimum number of samples required to be at a leaf node.
bootstrap = [True, False]#sampling 
```

```python
#dictionary for hyperparameters
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_clf1 = RandomForestClassifier(random_state=42)#model

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='f1',param_distributions=random_grid, 
                           n_iter=100, cv=3, 
                               verbose=2, random_state=42, n_jobs=-1)
```

```python
rf_cv.fit(X_train, y_train)##training data on randomsearch cv
rf_best_params = rf_cv.best_params_##it will give you best parameters 
print(f"Best paramters: {rf_best_params})")##printing  best parameters
```

```python
rf_clf2 = RandomForestClassifier(n_estimators=200,min_samples_leaf=4,min_samples_split=20,max_features="log2",max_depth=10)
#rf_clf2 = RandomForestClassifier(rf_best_params)#passing best parameter to randomforest
```
